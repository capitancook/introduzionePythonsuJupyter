# Funzioni e funzioni ricorsive
## le funzioni in Python

Vi ricordate "struttura dati + algoritmo = programma", qui siamo nella parte di algoritmo.
Immaginate di aver risolto brillantemente un problema applicativo con uno script (= programma) Python di vostra creazione.
Ad esempio, il seguente script per il calcolo della somma dei primi n numeri interi:


````python
n = int(input("Prego, inserire il numero di interi da sommare -->"))
sommaPrimiNNumeri = 0
for numero in range(1,n):
    sommaPrimiNNumeri = sommaPrimiNNumeri + numero
print( sommaPrimiNNumeri)
````

Per altro, come ripasso dei cicli in Python, lo stesso frammento di codice potevamo scriverlo come :

````python
n = int(input("Prego, inserire il numero di interi da sommare -->"))
somma = 0
i = 0
while i < n :
    somma = somma + i
    i = i + 1
print( somma)
````

Visto che avete creato uno script così ben congegnato non sarebbe male poterlo riutilizzare in modo semplice in altri script di vostra creazione dove una parte, magari minimale ma pur sempre una parte di un problema applicativo è il calcolo dei primi n numeri interi. Vero?
Questa cosa si può fare inserendo lo script che abbiamo realizzato in una funzione di Python.
Il procedimento è molto semplice.
Si prende il codice scritto e si individuano i valori di ingresso e i valori di uscita.
Nel nostro caso i valori di ingresso sono il numero di interi da sommare N mentre il valore di uscita o calcolato è la somma degli N interi.
I valori di ingresso diventeranno i parametri della funzione mentre i valori di uscita saranno quei valori che la funzione restituisce mediante l'istruzione return.
Inoltre, dobbiamo usare la fantasia o un metodo sistematico per dare il nome alla nuova funzione in modo che sia anche autoesplicativa.
Cosa ne dite di sommaNInteri ?
Per definire una nuova funzione si usa il comando def seguito dal nome della funzione seguito da parentesi tonda aperta, la lista dei parametri di ingresso separati dalla virgola (",") e dalla parentesi chiusa.
A seguire si aggiunge, spostato a destra di 4 caratteri, il blocco di istruzioni che abbiamo scritto in precedenza.
Vediamo come si scrive questa nuova funzione :

````python
def sommaNInteri(n):
    s = 0
    while n>0 :
       n = n-1
       s = s + n
    return s

sommaNInteri(3)
````

A questo punto l'interprete Python sa, se avete eseguito la cella precedente :), che esiste una nuovo funzione Python che si chiama sommaNInteri e quindi possiamo usarla nel nostro script. Ci penserà l'interprete Python a trasformare l'invocazione della funzione che andremo a fare in un qualche punto del nostro script con il codice contenuto nella funzione.
Ad esempio, possiamo replicare il funzionamento dello script originario con il seguente codice:

````python
n = int(input("Prego, inserire il numero di interi da sommare -->"))
print(sommaNInteri(n))
````

Oppure ancora più in breve:

````python
print(sommaNInteri(int(input("Prego, inserire il numero di interi da sommare -->"))))
````

Come vedete, una vpolta che avete definito una funzione ne potete fare uso in qualunque parte di uno script e l'interprete Python si occupa di esploderla nelle istruzioni che la compongono e eseguirle.
Ovviamente, perché l'interprete Python la conosca deve aver eseuito la sua def (inizione).


## Le procedure

Le funzioni possono restituire valori, tramite l'istruzione return come abbiamo visto, oppure possono non restituire valori ma svolgere solo dei compiti assegnati.
Ad esempio potremmo definirci una funzione o meglio una procedura (così si chiamano le funzioni quando non restituiscono valori :) denominata stampaCarta che stampa una carta da gioco di cui noi forniamo il seme e il valore come parametri di ingresso.

````python
def stampaCarta(seme,valore):
    print('┌─────────┐')
    print('│',valore,'      │')
    print('│         │')
    print('│   ',seme,'   │')
    print('│         │')
    print('│      ',valore,'│')
    print('└─────────┘')

stampaCarta('♥',"k")
````

# la ricorsione
Bene! Abbiamo visto che posso chiamare una funzione nel "corpo" di un'altra funzione. 
Ma cosa succede se in una funzione chiamo se stessa?
Succede che sto definendo qualcosa in funzione di se stessa ovvero sto usando una definizione ricorsiva.
In Python, e in molti altri linguaggi di programmazione si può fare. 
Vediamo insieme a cosa può servire.
Se riflettiamo bene ci sono molti problemi applicativi che possono essere trattati con la ricorsione.
Ad esempio, se vogliamo calcolare la somma dei primi N numeri interi possiamo osservare che questo problema è riconducibile alla somma di N-1 più l somma dei primi (N-1):
```
Sn = n-1 + Sn-1
```
Applicando ricorsivamente questa osservazione possiamo notare che la somma dei primi (N-1) numeri è uguale a (N-2) più la somma dei primi (N-3) numeri:
```
Sn = n-1 + (n-2) + Sn-2
```
E così via fino a 
```
Sn = n-1 + (n-2) + (n-3) + ... + n-(n-1) + S0
```
Ma, S0 lo sappiamo calcolare ed è pari a 0.
QUindi possiamo risalire alla somma dei primi N numeri osservando la sua definizione ricorsiva e il caso base che sappiamo risolvere per N=0.
Possiamo scrivere la funzione in uno pseudo Python nel seguente modo:
```
def sommaNNumeri(n):
    se n è uguale a 0 allora return 0
    altrimenti return n-1 + sommaNNumeri(n-1)
```
Traducendo lo pseudo linguaggio usato qui sopra in Python otteniamo:

````python
def sommaNNumeri(n):
    if n == 0: return 0
    return n - 1 + sommaNNumeri(n-1)

print(sommaNNumeri(3))
````

### Struttura di una funzione ricorsiva
Quindi se un problema applicativo e per sua natura ricorsivo possiamo usare l'approccio ricorsivo per la scrittura della funzione che ci aiuta a risolverlo.
La struttura della funzione sarà del tipo:
```
def f(n):
    if mi trovo nel caso base return valore base che so calcolare
    altrimenti applico la definizione ricorsiva del problema applicativo che ho individuato
```

La procedura che possiamo seguire per scrivere una funzione ricorsiva è:
+ individuare la struttura ricorsiva del problema;
+ individuare il caso base con la sua soluzione;
fatte queste due cose la funzione ricorsiva si scrive quasi da sola :). 



### Progetto di una funzione ricorsiva

Vogliamo ralizzare una funzione che stampa a video un traingolo isoscele retto composto da asterischi con lato di n asterischi.
Ad esempio, come questo di lato 5:
```
*
**
***
****
*****
```
Osservando il nostro problema possiamo notare che la base del tringolo e composta da n asterischi, il livello superiore è composto da n-1 asterischi e quello ancora superiore è composto da n-2 asterischi e così via fino all'apice che è composto da 1 asterisco e sopra ancora?
Non c'è nessun asterisco!
Quindi la definizione della figura da disegnare è ricorsiva e il caso base è semplice non devo stampare nulla.
L'implementazione in pseudocodice è la seguente:
```
def starTriangle(n):
    se n è uguale a 0 allora return nulla
    altrimenti 
        chiama starTriangle(n-1) per disegnare il piano di sopra 
        stampa il piano corrente return n-1 + sommaNNumeri(n-1)
```

````python
def starTriangle(n):
    if n == 0 :
        return
    else:
        starTriangle(n-1)
        print('*'*n)

starTriangle(10)
````
